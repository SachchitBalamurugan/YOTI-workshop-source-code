```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# Step 1: Create the dataset
data = {
    'Weather': ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Overcast'],
    'Temperature': ['Hot', 'Mild', 'Hot', 'Mild', 'Cold', 'Cold'],
    'Play': ['No', 'Yes', 'Yes', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

# Step 2: Convert categorical variables to numeric using one-hot encoding
df_encoded = pd.get_dummies(df[['Weather', 'Temperature']])

# Step 3: Separate features and target
X = df_encoded
y = df['Play']

# Step 4: Train a decision tree classifier
clf = DecisionTreeClassifier(criterion='entropy', random_state=0)
clf = clf.fit(X, y)

# Step 5: Plot the decision tree
plt.figure(figsize=(10,6))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True, rounded=True)
plt.title("Decision Tree Diagram")
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

data = {
    'Feature1': [1, 2, 3, 4, 5, 6],
    'Feature2': [0, 1, 0, 1, 0, 1],
    'Label':    [0, 1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)
X = df[['Feature1', 'Feature2']]
y = df['Label']

clf = DecisionTreeClassifier()
clf.fit(X, y)

fig = plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['Feature1', 'Feature2'], class_names=['No', 'Yes'], filled=True)
fig.savefig('decision_tree.png', dpi=300, bbox_inches='tight')
```
