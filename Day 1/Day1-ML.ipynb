# Linear Regression Model (STUDY DATA)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: Study hours vs Test scores
# X = number of study hours
# y = test score
X = np.array([[1], [2], [3], [4], [5]])  # feature (independent variable)
y = np.array([50, 55, 65, 70, 75])      # target (dependent variable)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predict scores for the same study hours
y_pred = model.predict(X)

# Print model coefficients
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")

# Plotting
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Predicted Line')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

# Linear Regression Model with Kaggle Dataset
#### https://www.kaggle.com/datasets/himanshunakrani/student-study-hours/data

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('Day 1/data.csv')

# Separate features and target
X = data[['Hours']]  # keep X as a 2D array
y = data['Scores']

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict using the trained model
y_pred = model.predict(X)

# Print the model parameters (the learned coefficients)
print("\n Linear Regression Equation:")
print("  y = m * x + b")
print(f"  Slope (m): {model.coef_[0]:.4f}")
print(f"  Intercept (b): {model.intercept_:.4f}")

# Calculate and print Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)
print("\n Mean Squared Error (MSE) Formula:")
print("  MSE = (1/n) * Σ(yᵢ - ŷᵢ)²")
print(f"  Computed MSE: {mse:.4f}")

# Display Gradient Descent update rules (theoretical explanation)
print("\n Gradient Descent Update Equations:")
print("  For each iteration:")
print("    m = m - α * (1/n) * Σ(-2xᵢ(yᵢ - (m*xᵢ + b)))")
print("    b = b - α * (1/n) * Σ(-2(yᵢ - (m*xᵢ + b)))")
print("  (Note: scikit-learn uses its own optimization internally)")

# Ask the user for study hours and predict test score
try:
    hours = float(input("\n Enter number of study hours to predict test score: "))
    predicted_score = model.predict([[hours]])[0]
    print(f" Predicted Test Score: {predicted_score:.2f}")
except ValueError:
    print(" Invalid input. Please enter a number.")

# Plot actual data and regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.title('Linear Regression (from CSV)')
plt.legend()
plt.show()
```
